```python
codon_table= {
		'ATA':'I', 'ATC':'I', 'ATT':'I', 'ATG':'M',
		'ACA':'T', 'ACC':'T', 'ACG':'T', 'ACT':'T',
		'AAC':'N', 'AAT':'N', 'AAA':'K', 'AAG':'K',
		'AGC':'S', 'AGT':'S', 'AGA':'R', 'AGG':'R',				
		'CTA':'L', 'CTC':'L', 'CTG':'L', 'CTT':'L',
		'CCA':'P', 'CCC':'P', 'CCG':'P', 'CCT':'P',
		'CAC':'H', 'CAT':'H', 'CAA':'Q', 'CAG':'Q',
		'CGA':'R', 'CGC':'R', 'CGG':'R', 'CGT':'R',
		'GTA':'V', 'GTC':'V', 'GTG':'V', 'GTT':'V',
		'GCA':'A', 'GCC':'A', 'GCG':'A', 'GCT':'A',
		'GAC':'D', 'GAT':'D', 'GAA':'E', 'GAG':'E',
		'GGA':'G', 'GGC':'G', 'GGG':'G', 'GGT':'G',
		'TCA':'S', 'TCC':'S', 'TCG':'S', 'TCT':'S',
		'TTC':'F', 'TTT':'F', 'TTA':'L', 'TTG':'L',
		'TAC':'Y', 'TAT':'Y', 'TGC':'C', 'TGT':'C',
		'TGG':'W', 'TAA':'' , 'TGA':'' , 'TAG':'' }
# replacing 'T's with 'U's in order to get mrna codons
rna_codons = codon_table.keys()
r_codons= []
for i in rna_codons:
    r_codons.append(i.replace('T','U'))
# passing on these updated codons as new keys to our codon_table dictionary
updated_codon = dict(zip(r_codons,list(codon_table.values())))
print(updated_codon)
```

```python
# reading the given text file
s = open('rosalind_prot(1).txt','r')
s= s.read()
s1= s.rstrip()
s1
```

```python
def translate(seq,g_codons:dict):
	protein = ""
	for i in range(0,len(seq),3):
		codon = seq[i:i+3]
		protein += g_codons[codon]
	return protein
translate(s1,updated_codon)
```
